```python
set1=562
set2=563

checkpoint = torch.load(f"saved_model2/model_{set1}.pt")
nn_model.load_state_dict(checkpoint['model_state_dict'])

checkpoint2 = torch.load(f"saved_model2/model_{set2}.pt")
nn_model2.load_state_dict(checkpoint2['model_state_dict'])
weightedLayerId= [0, 3, 6, 10, 12, 14, 16]
weightDiffers= []
for i in weightedLayerId:
    layer = i
    weightDiffer= nn_model.conv_pool_stack[layer].weight- nn_model2.conv_pool_stack[layer].weight
    weightDiffer= torch.sum(weightDiffer)
    weightDiffers.append(weightDiffer)

print(weightDiffers)
#nn_model2.conv_pool_stack[layer].weight
```

```python
def __pearsonr( x, y):
    cos = nn.CosineSimilarity(dim=0, eps=1e-6)
    """
    https://discuss.pytorch.org/t/use-pearson-correlation-coefficient-as-cost-function/8739
    :param x: the learned rbc vector
    :param y: the expected centrality vector
    :return pearson: the pearson corr score between the vectors
    """
    pearson = cos(x - x.mean(dim=0, keepdim=True), y - y.mean())
    return pearson
params = {'batch_size': 2,
        'shuffle': True,
        'collate_fn': LRCLoader.custom_collate_fn}
        
test_paths = [path[0] for path in list(os.walk("datas/Betweenness/test"))[1:]]

train_set = CustomDatasetLRCLoader(test_paths, TorchDtype.float, TorchDevice.gpu)
train_ldr = DataLoader(train_set, **params)
pearson_avg= 0
n_instances= len(train_ldr)
for i, samples in enumerate(train_ldr):
    Xs, ys = samples[0], samples[1]
    actual_rbcs = nn_model2(Xs)
    expected_rbcs= ys
    print(i)
    print(actual_rbcs)
    print(expected_rbcs)
    expected_rbcs[0] = expected_rbcs[0].to(TorchDevice.gpu)
    print(__pearsonr(actual_rbcs[0], expected_rbcs[0]))
```
